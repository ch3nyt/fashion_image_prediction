```python
import pandas as pd
sales = pd.read_csv("visuelle2/sales.csv")
print(sales.shape, sales.dtypes)
print(sales.head(5))
'''
external_code == image 

'''
```

```python
class image_info:
    def __init__(self, external_code, image_path, category, color, fabric):
        self.external_code = external_code
        self.image_path = image_path
        self.category = category
        self.color = color
        self.fabric = fabric
    
image_cat = {}
for _, row in sales.iterrows():
    image_cat[row["external_code"]] = image_info(
        row["external_code"],
        row["image_path"],
        row["category"],
        row["color"],
        row["fabric"],
    )

group_cols = ["external_code", "season", "release_date"]
sales_cols = [str(i) for i in range(12)]
agg_sales_subset = sales[group_cols + sales_cols]
agg_sales = agg_sales_subset.groupby(group_cols).sum().reset_index()
```

```python
cross_season = []
for i in range(1, 5578):
    if len(agg_sales[agg_sales["external_code"] == i]) > 5:
        cross_season.append(i)
print(cross_season)
```

```python
agg_sales[agg_sales["external_code"].isin(cross_season)].tail(30)
```

```python
# 1 code   path
#  code  path
code_to_path_counts = sales.groupby('external_code')['season'].nunique()

# .gt(1)  1  (True/False).sum()  True
code_violations = code_to_path_counts.gt(1).sum()

if code_violations > 0:
    print(f" {code_violations}  external_code  season")
else:
    print("1 external_code  season")

# 2 path   code
#  path  code
path_to_code_counts = sales.groupby('season')['external_code'].nunique()

#  1 
path_violations = path_to_code_counts.gt(1).sum()

if path_violations > 0:
    print(f" {path_violations}  season  external_code")
else:
    print("2 season  external_code")

# 
if code_violations == 0 and path_violations == 0:
    print("\n")
else:
    print("\n")
```
